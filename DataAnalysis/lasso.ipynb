```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler

# Enable inline plotting in Jupyter Notebook
%matplotlib inline

# Read the CSV file
data = pd.read_csv('NONseasonticket.csv')

# Define predictor columns
predictors = ['FanSegment', 'DistanceToArena', 'BasketballPropensity', 'SocialMediaEngagement']

# Extract predictor data
X = data[predictors].copy()

# Convert numeric columns if needed
X['DistanceToArena'] = pd.to_numeric(X['DistanceToArena'], errors='coerce')
X['BasketballPropensity'] = pd.to_numeric(X['BasketballPropensity'], errors='coerce')

# One-hot encode the categorical predictors: FanSegment and SocialMediaEngagement
X = pd.get_dummies(X, columns=['FanSegment', 'SocialMediaEngagement'], drop_first=True)

# Fill missing values in predictors with column means
X.fillna(X.mean(), inplace=True)
```

```python
# Define target column
target = "A Games"

# Convert target column to numeric and filter out missing values
y = pd.to_numeric(data[target], errors='coerce')
valid_idx = y.notnull()
X_valid = X[valid_idx]
y_valid = y[valid_idx]

# Standardize predictors on the valid subset
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_valid)

# Initialize and fit the Lasso model (adjust alpha as needed)
model = Lasso(alpha=0.1)
model.fit(X_scaled, y_valid)
y_pred = model.predict(X_scaled)

# Plot actual vs. predicted values
plt.figure(figsize=(8, 6))
plt.scatter(y_valid, y_pred, alpha=0.7, edgecolor='k')
plt.xlabel('Actual ' + target)
plt.ylabel('Predicted ' + target)
plt.title('Lasso Regression for ' + target)
min_val = min(y_valid.min(), y_pred.min())
max_val = max(y_valid.max(), y_pred.max())
plt.plot([min_val, max_val], [min_val, max_val], 'r--')
plt.show()

# Print Lasso regression coefficients (for standardized predictors) and intercept
print("Lasso Regression coefficients for", target, ":")
print(pd.Series(model.coef_, index=X.columns))
print("Intercept:", model.intercept_)
```
